# Binning ICESat-2 Freeboards in CICE5 Model Grid Cells

### Objective

Sample and write ICESat-2 observations that will play nicely with assimilation into CICE5. The main point is to reduce the obs in a way that maximizes our retention of information 

Primary options for data reduction:
- take a mean or median of cell-aggregated satellite measurements
- fit the data to the model's ice thickness bins, thereby reproducing the ITD

Important pieces of information needed to write the data back into a pre-DART observation file:
- the reduced measurement
- the total observational uncertainty (satellite error + std of the distribution, to start)
- the location (latitude and longitude)

### Setup

```python
import geopandas as gpd
import xarray as xr
from shapely.geometry import Polygon
from shapely.geometry import Point
import numpy as np
import matplotlib.pyplot as plt
import glob
```

**First, open grid and satellite data.**

```python
# open grid data
grid = xr.open_dataset('../cice_grid.nc')

# open a single satellite track
ds1 = xr.open_dataset('/glade/scratch/mollyw/external_data/IS2_reduced/IS2_2019.01.15.nc')

# open all satellite tracks
files = sorted(glob.glob('/glade/scratch/mollyw/external_data/IS2_reduced/IS2_2019.01.*[0-9]*.nc'))
ds = xr.open_mfdataset(files, combine='nested', concat_dim='time')
```

**Next, form Geometries.**

```python
# convert grid cells into Polygons

# initalize empty lists for the cell data and the thickness data
grid_cells = []
grid_hi = []

# cycle through all grid indices ...
for x in range(0, np.shape(grid.TLAT)[0]):
    for y in range(0, np.shape(grid.TLON)[1]):

        # collecting the latitude (y) and longitude (x) of each of the vertices
        x0 = grid.lont_bounds[x, y, 0].values
        x1 = grid.lont_bounds[x, y, 1].values
        x2 = grid.lont_bounds[x, y, 2].values
        x3 = grid.lont_bounds[x, y, 3].values
        y0 = grid.latt_bounds[x, y, 0].values
        y1 = grid.latt_bounds[x, y, 1].values
        y2 = grid.latt_bounds[x, y, 2].values
        y3 = grid.latt_bounds[x, y, 3].values

        # if any of the grid corner data is NaN, pass on it
        if sum(np.isnan([x0, x1, x2, x3, y0, y1, y2, y3])) > 0:
            pass

        # otherwise, generate a Polygon from the corners and store it in the
        # list along with the thickness value of the grid cell
        else:
            # print(x0, x1, y1, y0)
            grid_cells.append(Polygon([(x0, y0), (x1, y1), (x2, y2), (x3, y3)]))
            grid_hi.append(grid.hi[0, x, y].values.item())

# turn the Polygons into a GeoDataFrame with the standard WGS84 CRS
cell = gpd.GeoDataFrame(grid_cells, columns=['geometry'], crs='epsg:4326')

# add in the thickness values
cell["thickness"] = grid_hi

# convert the CRS to a Northern Hemisphere Polar Stereographic projection
cell.to_crs('epsg:3995', inplace=True)

# if you'd like, look at the output
print(cell.head())
```

Note: as of 06/09/2022, the thickness values of the file I'm using for the model grid are pretty bogus. The file is the first monthly output of a control run, and thus it's thickness values are not anything close to realistic. I'm including the make sure that I can get that input in the future and to help visualize what's happening when I plot the grid for sanity checks.

```python
# convert satellite data to Points

# convert DataArrays to DataFrames (pandas)
df1 = ds1.to_dataframe().reset_index()
df = ds.to_dataframe().reset_index()

# generate Points with an iterator
points1 = [Point(x, y) for x, y in zip(df1['longitude'], df1['latitude'])]
points = [Point(x, y) for x, y in zip(df['longitude'], df['latitude'])]

# convert to a GeoDataFrame
satellite_track = gpd.GeoDataFrame(df1, geometry=points1, crs='epsg:4326')
satellite = gpd.GeoDataFrame(df, geometry=points, crs='epsg:4326')

# convert the CRS to the same as the grid
satellite_track.to_crs(cell.crs, inplace=True)
satellite.to_crs(cell.crs, inplace=True)

# if you'd like, look at the output
print(satellite.head())
```

Note: For the single track, I'm using ICESat-2 measurements from January 15th, 2019. For the full field, I'm using all the ICESat-2 measurements taken in the month of January, 2019. We should expect there to be biases introduced by using an entire month's worth of data as a single field without accounting for the evolution that would occur in place over that month. 

### Analysis

**Start by binning the satellite data by grid cell.**

```python
# use GeoPandas "spatial join" to aggregate the satellite points by the grid
# cells that contain them.
binned_track = gpd.sjoin(cell, satellite_track, how='left', op='contains')
binned = gpd.sjoin(cell, satellite, how='left', op='contains')

# drop any rows that are all NaNs/NaTs
binned_track = binned_track.dropna()
binned = binned.dropna()

# pull and assign each bin a "cell_number". This is identical to the index, but
# accessible to operate over for grouping functions
track_index_list = list(binned_track.index.values)
binned_track["cell_number"] = track_index_list
index_list = list(binned.index.values)
binned["cell_number"] = index_list
```

**Now reduce the data using mean, median, and distribution.**

```python
# do the easy ones first, the mean and the median

# perform a description of the data on a groupby object
track_mean = binned_track.dissolve(by='cell_number', aggfunc='mean')
mean = binned.dissolve(by='cell_number', aggfunc='mean')

track_median = binned_track.dissolve(by='cell_number', aggfunc='mean')
median = binned.dissolve(by='cell_number', aggfunc='mean')

# STILL TO DO: FIND A GOOD WAY OF ISOLATING NEEDED INFO AND WRITING TO FILE
```

```python
# now do the distribution

# group by cell_number and aggregate in lists based on index
binned_track = binned_track.groupby('cell_number').agg(lambda x: list(x))
binned = binned.groupby('cell_number').agg(lambda x: list(x))

# determine the number of freeboard values per index
counts1 = [len(x) for x in binned_track["freeboard"]]
counts = [len(x) for x in binned["freeboard"]]

# save counts as part of the GeoDataFrame
binned_track["count"] = counts1
binned["count"] = counts

# filter out cells that do not have sufficent satellite returns
binned_track = binned_track.where(binned_track["count"] > 10).dropna()
binned = binned.where(binned["count"] > 100).dropna()
```
